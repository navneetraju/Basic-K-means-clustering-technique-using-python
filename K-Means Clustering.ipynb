```python
import random
import csv
```

```python
ages=[15,20,34,12,15,20,25,67,78,89,90,34,45,56,71]
with open('C:/Users/Navneet/Desktop/height-weight.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    heights = []
    weights = []
    for row in readCSV:
        height = row[0]
        weight = row[1]

        heights.append(height)
        weights.append(weight)
for i in range(0,len(heights)):
    heights[i]=float(heights[i])
```

```python
def k_means(data,k):
    means=random.sample(data,k)
    #mean_list=list(k)
    #print(means)
    def kmeans(data,means):
        clusters=[[] for a in means]
        #print(clusters)
        for d in data:
            distances=[(abs(d-m),i) for i,m in enumerate(means)]
            #print(distances)
            a,closest=min(distances)
            #print(closest)
            clusters[closest].append(d)
        return sorted(clusters)
    prev=[]
    curr=kmeans(data,means)
    #print(curr)
    while curr!=prev:
        prev=curr
        means=[sum(c)/max(1,len(c)) for c in curr]
        curr=kmeans(data,means)
    return zip(means,curr)
```
